```python
import logging

# 1. Configure the logging system
logging.basicConfig(
    level=logging.info,                    # The minimum severity level to capture
    format="%(asctime)s [%(levelname)s] %(message)s",  # How the log line looks
    handlers=[
        logging.FileHandler("app.log"),    # Saves logs to a file
        logging.StreamHandler()            # Also prints logs to your console
    ]
)
logging.info("this is the starting point")
logging.warning("this is the worning")
logging.debug("this is the worning")
```

```python
import logging

logger=logging.getLogger(__name__)
logger.basicConfig(
    level=logging.info,                    # The minimum severity level to capture
    format="%(asctime)s [%(levelname)s] %(message)s",  # How the log line looks
    handlers=[
        logging.FileHandler("app.log"),    # Saves logs to a file
        logging.StreamHandler()            # Also prints logs to your console
    ]
)
logger.basicConfig(
    level=logging.warn,                    # The minimum severity level to capture
    format="%(asctime)s [%(levelname)s] %(message)s",  # How the log line looks
    handlers=[
        logging.FileHandler("app.log"),    # Saves logs to a file
        logging.StreamHandler()            # Also prints logs to your console
    ]
)
logger.basicConfig(
    level=logging.DEBUG,                    # The minimum severity level to capture
    format="%(asctime)s [%(levelname)s] %(message)s",  # How the log line looks
    handlers=[
        logging.FileHandler("app.log"),    # Saves logs to a file
        logging.StreamHandler()            # Also prints logs to your console
    ]
)
logger.basicConfig(
    level=logging.ERROR,                    # The minimum severity level to capture
    format="%(asctime)s [%(levelname)s] %(message)s",  # How the log line looks
    handlers=[
        logging.FileHandler("app.log"),    # Saves logs to a file
        logging.StreamHandler()            # Also prints logs to your console
    ]
)
```
