```python
import whisper
from langchain_core.document_loaders import BaseLoader
from langchain_core.documents import Document

class LocalVideoWhisperLoader(BaseLoader):
    """A custom LangChain loader to extract text from any local video file using Whisper."""
    def __init__(self, video_path: str, model_name: str = "base"):
        self.video_path = video_path
        self.model_name = model_name

    def load(self):
        
        model = whisper.load_model(self.model_name)

        
        print(f"Transcribing {self.video_path} locally...")
        result = model.transcribe(self.video_path)
        

        
        return [
            Document(
                page_content=result["text"],
                metadata={"source": self.video_path, "language": result.get("language")}
            )
        ]



loader = LocalVideoWhisperLoader(video_path="/home/mikealson/Videos/Spring_Boot_ API_Request_Validation_&_Exception_Handling_Realtime_Exaple_with_JavaTechie.mkv", model_name="base")
docs = loader.load()


print(docs[0].page_content)
```
