```python
import os
from openai import OpenAI
from dotenv import load_dotenv 
load_dotenv()

client = OpenAI(
    base_url="https://router.huggingface.co/v1",
    api_key=os.environ["HUGGINGFACE_API_KEY"], 
)

user_input=input("what do you think:") #use input


completion = client.chat.completions.create( #the primary command used to send a message to an AI model and get a reply back.
    model="deepseek-ai/DeepSeek-V4-Pro:novita", #hugging face model
    messages=[
         {
            "role": "system", 
            "content": "You are a helpful assistant that speaks like a pirate."  # Your system prompt goes here
        },
        {
            "role": "user",
            "content": user_input   #user input comes here and goes to the model passed to the hugging face model
        }
    ],
)

response =completion.choices[0].message #This line of Python code is used to grab the final text answer sent back by an AI model.
text_output = response.content #this will give you the response in text format
print(text_output)
```

```python
#!pip install -qU Groq 
from openai import OpenAI
import os

from dotenv  import load_dotenv
load_dotenv()
grok_api_key=os.environ.get("Groq_api_key")
client = OpenAI(
    api_key=grok_api_key,
    base_url="https://api.groq.com/openai/v1",
)
user_input="" #initializing the empty input to get enter in the loop
while (user_input!="exit"):  #using the loop to ask the input multiple times
    user_input=input("what do you think: ")
    response = client.responses.create( # it is necessary to ge the response from the user
        input=user_input,
        model="openai/gpt-oss-20b",
    )
    output_data=response.output_text
    print(output_data)
```
