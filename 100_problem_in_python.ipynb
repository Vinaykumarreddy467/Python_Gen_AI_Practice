```python
# 16. Write a Python program to generate the Fibonacci series up to n terms.
def fib(n):
    l1 = []
    if n <= 0:
        return l1
    elif n == 1:
        l1.append(0)
    else:
        n1, n2 = 0, 1
        l1.extend([n1, n2])
        for i in range(2, n):
            nth = n1 + n2
            l1.append(nth)
            n1 = n2
            n2 = nth
    return l1

num =10
l2=fib(num)
print(f'''Fibonacci series up to {num} terms: {l2}''')

```

```python
# 23. Write a Python program to check whether a number is an Armstrong number.
num=int(input("enter a number to check armstrong or not: "))
str_num=str(num)
len_num=len(str(num))
total=0
for i in str_num : total+=int(i)**len_num 
if total== num : print("Armstrong number")
else : print("not a Armstrong number")
```
