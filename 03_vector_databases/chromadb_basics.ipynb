```python
import pypdf

def extract_text(file_path: str) -> str:
    """Extracts text from a PDF file using pypdf."""
    text_parts = []
    
    with open(file_path, "rb") as f:
        reader = pypdf.PdfReader(f)
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                text_parts.append(page_text)
                
    # Joining at the end is much faster than repeated '+' concatenation
    return text_parts

# Execution
docs = extract_text("/home/mikealson/Documents/Vinay_Kumar_Reddy_Frontend_FullStack_Developer.pdf")

metadatas=[]
def ides(docs):
    idlist=[]
    
    for i in range(0,len(docs)):
        idlist.append(str(i))
        metadatas.append(f"this is my resume part{i}")
    return idlist
```
