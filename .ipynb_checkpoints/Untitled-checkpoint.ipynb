```python
#Create a new post using POST.
url="https://jsonplaceholder.typicode.com/posts"
res=requests.get(url).json()
print(res[1])
```

```python
#Update a post using PUT/PATCH.
new_post = {
    'userId': 10, 
    'id': 1, 
    'title': 'sunt aut facere repellat provident occaecati excepturi optio reprehenderit', 
    'body': 'quia et suscipit\nsuscipit recusandae consequuntur expedita et cum\nreprehenderit molestiae ut ut quas totam\nnostrum rerum est autem sunt rem eveniet architecto'
}

# 1. Fix the URL by targetting the specific resource id (/posts/1)
# 2. Pass the dictionary to the 'json' argument so it sends as valid JSON text
res = requests.put(f"{url}/{new_post['id']}", json=new_post)

print(res.status_code)
```
