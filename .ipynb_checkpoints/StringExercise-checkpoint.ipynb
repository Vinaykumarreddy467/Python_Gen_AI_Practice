```python
# ==============================================================================
# PART 1: BASIC STRING MANIPULATIONS (EXPRESSIONS & OUTPUTS)
# ==============================================================================

# 1. Length of "Artificial Intelligence"
print(f'1. Length: {len("Artificial Intelligence")}')

# 2. Convert "python programming" into uppercase
print(f'2. Uppercase: {"python programming".upper()}')

# 3. Convert "MACHINE LEARNING" into lowercase
print(f'3. Lowercase: {"MACHINE LEARNING".lower()}')

# 4. Convert "hello world" into title case
print(f'4. Title case: {"hello world".title()}')

# 5. Capitalize the first letter of "python"
print(f'5. Capitalize: {"python".capitalize()}')

# 6. Remove extra spaces from " Data Science "
print(f'6. Remove spaces: {" Data Science ".replace(" ", "")}')

# 7. Remove only left-side spaces from " Python"
print(f'7. Left strip: {" Python".lstrip()}')

# 8. Remove only right-side spaces from "Python "
print(f'8. Right strip: {"Python ".rstrip()}')

# 9. Find the position of "thon" in "Python"
print(f'9. Position of "thon": {"Python".find("thon")}')

# 10. Find the position of "AI" in "Generative AI"
print(f'10. Position of "AI": {"Generative AI".find("AI")}')

# 11. What happens when index() searches for a missing substring?
try:
    "Python".index("Java")
except ValueError as e:
    print(f'11. index() on missing substring raises Exception: {type(e).__name__} ({e})')

# 12. Find the first occurrence of "a" in "banana"
print(f'12. Find "a": {"banana".find("a")}')

# 13. Replace "Python" with "Java" in "I love Python"
print(f'13. Replace word: {"I love Python".replace("Python", "Java")}')

# 14. Replace all spaces with underscores in "Machine Learning"
print(f'14. Replace spaces with underscores: {"Machine Learning".replace(" ", "_")}')

# 15. Replace "2025" with "2026" in "Year 2025"
print(f'15. Replace year: {"Year 2025".replace("2025", "2026")}')

# 16. Split "apple,banana,mango" into a list
print(f'16. Split by comma: {"apple,banana,mango".split(",")}')

# 17. Split "Python AI ML DataScience" based on spaces
print(f'17. Split by space: {"Python AI ML DataScience".split()}')

# 18. Join ["Python","AI","ML"] using a hyphen (-)
print(f'18. Join with hyphen: {"-".join(["Python", "AI", "ML"])}')

# 19. Join ["2025","06","01"] to form a date
print(f'19. Join to form date: {"-".join(["2025", "06", "01"])}')

# 20. Count the number of "a" characters in "banana"
print(f'20. Count "a": {"banana".count("a")}')

# 21. Count the occurrences of "is" in "This is a test. This is Python."
print(f'21. Count "is": {"This is a test. This is Python.".count("is")}')

# 22. Count how many spaces are present in a sentence
print(f'22. Count spaces: {"Count how many spaces".count(" ")}')

# 23. Check whether "data.csv" ends with .csv
print(f'23. Ends with .csv: {"data.csv".endswith(".csv")}')

# 24. Check whether "https://google.com" starts with https
print(f'24. Starts with https: {"https://google.com".startswith("https")}')

# 25. Check whether "python.py" ends with .py
print(f'25. Ends with .py: {"python.py".endswith(".py")}')

# 26. Check whether "Python" contains only alphabets
print(f'26. Is alpha: {"Python".isalpha()}')

# 27. Check whether "12345" contains only digits
print(f'27. Is digit: {"12345".isdigit()}')

# 28. Check whether "Python123" is alphanumeric
print(f'28. Is alnum: {"Python123".isalnum()}')

# 29. Check whether " " contains only spaces
print(f'29. Is space: {" ".isspace()}')

# 30. Check whether "Python@" is alphanumeric
print(f'30. Is alnum (with @): {"Python@".isalnum()}')


# ==============================================================================
# PART 2: LOGIC BLOCKS & ALGORITHMS
# ==============================================================================

print("\n" + "="*50 + "\nPART 2: LOGIC BLOCKS & ALGORITHMS\n" + "="*50)

# 1. Count vowels in a string
vowel_text = "artificial intelligence"
vowel_count = sum(1 for char in vowel_text.lower() if char in "aeiou")
print(f"1. Vowels in '{vowel_text}': {vowel_count}")

# 2. Reverse a string without using slicing
reverse_text = "Python"
reversed_str = "".join(reversed(reverse_text))
print(f"2. Reversed '{reverse_text}': {reversed_str}")

# 3. Check whether a string is a palindrome
palindrome = "Radar"
clean_text = "".join(palindrome.split()).lower()
is_palindrome = clean_text == clean_text[::-1]
print(f"3. Is '{palindrome_text}' a palindrome?: {is_palindrome}")

# 4. Count the number of words in a sentence
sentence_text = "Python AI ML DataScience"
word_count = len(sentence_text.split())
print(f"4. Word count in '{sentence_text}': {word_count}")

# 5. Find the frequency of each character in a string
freq_text = "banana"
frequency = {}
for char in freq_text:
    frequency[char] = frequency.get(char, 0) + 1
print(f"5. Character frequency in '{freq_text}': {frequency}")

# 6. Remove all spaces from a string
spaced_text = "  Machine   Learning  "
no_spaces = spaced_text.replace(" ", "")
print(f"6. Removed spaces from '{spaced_text}': '{no_spaces}'")

# 7. Convert a sentence into title case
lower_sentence = "this is a python tutorial."
title_sentence = lower_sentence.title()
print(f"7. Title case sentence: {title_sentence}")

# 8. Find the longest word in a sentence
long_word_sentence = "Find the longest word in a sentence"
words = long_word_sentence.split()
longest = max(words, key=len)
print(f"8. Longest word in '{long_word_sentence}': '{longest}'")

# 9. Check whether two strings are anagrams
anagram_1 = "listen"
anagram_2 = "silent"
is_anagram = sorted(anagram_1.lower()) == sorted(anagram_2.lower())
print(f"9. Are '{anagram_1}' and '{anagram_2}' anagrams?: {is_anagram}")

# 10. Remove duplicate characters from a string
dup_text = "banana"
unique_text = "".join(dict.fromkeys(dup_text))
print(f"10. Duplicates removed from '{dup_text}': '{unique_text}'")
```
