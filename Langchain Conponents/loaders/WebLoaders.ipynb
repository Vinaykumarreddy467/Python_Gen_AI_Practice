```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv
from langchain_core.output_parsers import StrOutputParser
from langchain_community.document_loaders import (
    TextLoader, 
    PyPDFLoader, 
    WebBaseLoader, 
    CSVLoader, 
    DirectoryLoader
)
from langchain_core.prompts import PromptTemplate

load_dotenv()
llm= ChatGroq(
    model="llama-3.3-70b-versatile", # You can also use mixtral-8x7b-32768
    temperature=0.5
)

```

```python
prompt = PromptTemplate(
    template="Write a summary for the following content from the website:\n{content}", 
    input_variables=['content']
)
```

```python
parser=StrOutputParser()
```

```python
url='https://surrealdb.com/learn/tour/page-02'
```

```python
loader=WebBaseLoader(url)
```

```python
docs=loader.load()
```

```python
print(docs[:500])
```

```python
print(docs[0])
```

```python
all_metadata = [doc.metadata for doc in docs]

# Example: Print the source URL for each document
for meta in all_metadata:
    print(meta.get('description'))
```

```python
chain=prompt|llm|parser
```

```python
res=chain.invoke([docs[0].page_content])
```
