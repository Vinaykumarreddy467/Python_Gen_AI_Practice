```python
import os 
import logging
from openai import OpenAI
from dotenv import load_dotenv 
load_dotenv()


logger = logging.getLogger(__name__)

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    handlers=[
        logging.FileHandler("app.log")
    ]
    
)
logging.basicConfig(
    level=logging.ERROR,
    format='%(asctime)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    handelers=[
        logging.fileHandler("app.log")
    ]
)
logging.basicConfig(
    level=logging.debug,
    format='%(asctime)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
logging.basicConfig(
    level=logging.warning,
    format='%(asctime)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    handelers=[
        logging.fileHandler("app.log")
    ]
)

logger.info("logger and environment variables lodded successfully")


# 1. Set up the client
client = OpenAI(
    base_url="https://router.huggingface.co/v1",
    api_key=os.environ["HUGGINGFACE_API_KEY"], 
)
logger.info("Hugging Face connected successfully")

# 2. Create the list of inputs you want to send
# (Using different questions so you get different answers!)
inputs = [
    "Hello there!",
    "Where is the treasure?",
    "How do I steer this ship?",
    "What is for dinner?"
]
responses = []
logger.info("Starting to get responses from the model...")

for user_input in inputs:
    # Use an active text/chat model on the Hugging Face router
    completion = client.chat.completions.create(
        model="Qwen/Qwen2.5-7B-Instruct", 
        messages=[
            {"role": "system", "content": "You are a helpful assistant that speaks like a pirate."},
            {"role": "user", "content": user_input}
        ]
    )
    # 🛑 SAFEGUARD: Check if the response returned a string error instead of an object
    if isinstance(completion, str):
       break
    # Extract the text answer properly 
    ans = completion.choices[0].message.content
    responses.append(ans)

# 4. Print all retrieved answers
for i, res in enumerate(responses, 1):
    print(f"{i}. Response = {res}")
```

```python
import asyncio
from openai import AsyncOpenAI

# 1. Set up the ASYNC client pointing to Ollama
client = AsyncOpenAI(
    base_url="http://localhost:11434/v1",  
    api_key="ollama",                      
)

inputs = [
    "Hello there!",
    "Where is the treasure?",
    "How do I steer this ship?",
    "What is for dinner?"
]

# 2. Define a function for a single streaming task
async def stream_pirate_response(batch_id, user_input):
    completion_stream = await client.chat.completions.create(
        model="tinyllama:latest",               
        messages=[
            {"role": "system", "content": "You are a helpful assistant that speaks like a pirate."},
            {"role": "user", "content": user_input}
        ],
        stream=True  
    )
    
    async for chunk in completion_stream:
        if chunk.choices[0].delta.content is not None:
            text_piece = chunk.choices[0].delta.content
            print(f"{text_piece}", end="", flush=True)

# 3. Main function to bundle and run the batch
async def main():
    print("🚀 Starting the async batch processing...\n")
    
    tasks = []
    for i, user_input in enumerate(inputs, 1):
        task = asyncio.create_task(stream_pirate_response(i, user_input))
        tasks.append(task)
        
    await asyncio.gather(*tasks)
    print("\n\n✅ All batch items finished processing!")

# 4. FIX FOR JUPYTER: Use top-level await instead of asyncio.run()
await main()
```
