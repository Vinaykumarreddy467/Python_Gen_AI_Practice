**What is JSON?**

JSON = JavaScript Object Notation

It is a lightweight text format for storing and exchanging data. Although it comes from JavaScript, it is language-independent – almost every programming language can read/write JSON.

Analogy: Think of JSON as a universal shipping box. You can pack your data (toys, books, clothes) inside it, and anyone anywhere (Python, Java, C#) can unpack it correctly. The box has clear labels ("name", "age") so nothing gets lost.


**Why is JSON used?**

Human-readable – easy to understand and edit

Machine-friendly – simple to parse and generate

Lightweight – less verbose than XML

Language-independent – works with Python, Java, PHP, etc.

Native web support – browsers understand it directly



Where is JSON used in real life?

Industry  	Web APIs	:- REST APIs (Twitter, Google Maps, YouTube)
Mobile apps	:- Data exchange between server and app
Configuration files :- 	package.json (Node.js), .vscode/settings.json
IoT devices :- 	Sensor data transmission
Databases :- 	PostgreSQL (JSONB), MongoDB (BSON)
Data pipelines:- Logging


**Benefits of JSON over XML**:

Smaller size (less bandwidth)

Faster to parse

Easier to read

Native array support

Direct mapping to programming objects

XML (eXtensible Markup Language)

XML is a markup language used to store, transport, and organize data in a structured, human-readable format.

Think of XML as a way to describe data using custom tags.

Why XML?
Stores data in a structured format
Easy for humans and machines to read
Platform-independent
Widely used for data exchange between systems
Used in configuration files, web services (SOAP), and document storage

3. jmespath Library (Query Language for JSON)


```
! pip install jmespath
```
What it does: Query nested JSON like SQL queries a database

Why use it: Instead of writing complex nested loops, write a query string

Without jmespath (verbose):



```
# This is formatted as code
# Get all user emails where user is active and age > 25
emails = []
for user in data["users"]:
    if user.get("active") and user.get("age", 0) > 25:
        emails.append(user.get("email"))
```

With jmespath (concise):


```
import jmespath
emails = jmespath.search("users[?active==`true` && age>`25`].email", data)
```
Common JMESPath Expressions:



```
# This is formatted as code
# Basic navigation
jmespath.search("user.name", data)           # Get single value
jmespath.search("users[0].name", data)       # First user's name
jmespath.search("users[*].name", data)       # All user names

# Filtering
jmespath.search("users[?age>30]", data)              # Users over 30
jmespath.search("users[?city=='NYC'].name", data)    # NYC users' names

# Projection
jmespath.search("users[*].{full_name: name, city: city}", data)

# Functions
jmespath.search("length(users)", data)               # Count users
jmespath.search("max_by(users, &age).name", data)    # Oldest user
```

What it does: Safe, declarative way to extract deeply nested data


Why use it: Handles missing keys gracefully and reduces try-except blocks


Without glom (error-prone):



```

    email = "Not found"

```

#With glom

```

email = glom(data, "user.profile.contact.email", default="Not found")
```

















````python
# Advanced glom features
!pip install glom
from glom import glom, Coalesce, Iter

# Define a sample 'data' dictionary with a 'users' key for glom examples
data = {
    "user": {"email": "user@example.com"},
    "profile": {"email": "profile@example.com"},
    "contact": {"email": "contact@example.com"},
    "users": [
        {"name": "Alice", "age": 30, "active": True, "email": "alice@example.com"},
        {"name": "Bob", "age": 25, "active": False, "email": "bob@example.com"},
        {"name": "Charlie", "age": 35, "active": True, "email": "charlie@example.com"}
    ]
}

# Handle multiple possible paths
spec = Coalesce("user.email", "profile.email", "contact.email")
email = glom(data, spec, default="No email")
print(f"Email (Coalesce): {email}")

# Extract from all items in list (fixed: changed to extract name and age into a dictionary for each user)
spec = ("users", [{"name": "name", "age": "age"}])  # Get name and age for all users
results = glom(data, spec)
print(f"Users (name, age): {results}")

# Skip missing items
spec = ("users", Iter().filter(lambda x: x.get("active")).map(lambda x:x["name"]).all())
active_names = glom(data, spec)
print(f"Active user names: {active_names}")
````

5. jsonpath-ng Library (XPath for JSON)
*`pip install jsonpath-ng`*
What it does: Implements JSONPath - a query language similar to XPath

Why use it: Most powerful querying for complex nested structures

Basic Usage:


```
# This is formatted as code
from jsonpath_ng import parse

# Compile expression once
jsonpath_expr = parse("$.store.book[*].title")

# Find all matches
matches = jsonpath_expr.find(data)
titles = [match.value for match in matches]
```


```
# This is formatted as code
# Root object
expr = parse("$")                    # Entire data
expr = parse("$.store")              # Store object

# Child operators
expr = parse("$.store.book")         # Dot notation
expr = parse("$['store']['book']")   # Bracket notation

# Wildcards
expr = parse("$.store.book[*]")      # All books
expr = parse("$.store.*")            # Everything in store

# Recursive descent
expr = parse("$..price")             # All prices anywhere
expr = parse("$..book[0]")           # First book anywhere

# Filter expressions
expr = parse("$.store.book[?(@.price<10)]")        # Books under $10
expr = parse("$.store.book[?(@.category=='fiction')]")  # Fiction books

# Slicing
expr = parse("$.store.book[0:2]")    # First two books
expr = parse("$.store.book[-2:]")    # Last two books
```

Core JSON Terms

Serialization	=> Converting Python object → JSON string =>	json.dumps()

Deserialization	=> Converting JSON string → Python object	=> json.loads()

Parsing	 => Process of analyzing JSON structure	 => response.json()

Nested	=> Objects or arrays inside other objects	 => {"user": {"name": "John"}}

Key-Value Pair =>  	A key mapped to a value	=> "name": "John"

Root	=> Top-most level of JSON	=> Everything inside {} or []



API (Application Programming Interface)=>	A set of rules for software to communicate

REST API =>	API that follows REST architectural style (most common for JSON)

Endpoint =>	Specific URL where API can be accessed

HTTP Method	=>Action to perform: GET (read), POST (create), PUT (update), DELETE (remove)

Status Code =>	3-digit number indicating result: 200=OK, 404=Not Found, 500=Server Error

Request	 => Message sent to API

Response =>	Data returned from API (often JSON)

Payload	=>Actual data sent/received (not including headers/metadata)

Header =>	Metadata about request/response (auth tokens, content type)

Query Parameter	 => Data appended to URL: ?key=value&page=1

Error Handling Terms

Exception =>	Error that occurs during execution

KeyError=>	When dictionary key doesn't exist

IndexError =>	When list index is out of range

JSONDecodeError	=> When string is not valid JSON

HTTPError	=>When API returns error status code (4xx, 5xx)

Timeout =>	When request takes too long

Graceful Degradation	=> Providing default values instead of crashing

Defensive Programming	=> Assuming data might be missing/invalid
