```python
import requests
import os
from dotenv import load_dotenv
  
load_dotenv()

dog_api_key=os.getenv("DOG_API_KEY")
print(dog_api_key)
```

```python
url="https://api.thedogapi.com/v1/"
headers = {
    "x-api-key":  dog_api_key
}
response=requests.get(url+"breeds",  headers=headers).json()
print(response[0])
```

```python
response=requests.get(url+"breeds/search?q=terrier",  headers=headers).json()
# print(contentType = response.headers.get('Content-Type'))
print(response[0])
```

```python

response=requests.get(url+"breeds?breed_groups=Toy,Working",  headers=headers)
res=response.json()
print(response.status_code)
# print(response)
print(res[0])
print(res[0]["weight"])
```

```python

response=requests.get(url+"pets/{1}/body-condition-score",  headers=headers)
res=response.json()
print(response.status_code)
print(res)
```
