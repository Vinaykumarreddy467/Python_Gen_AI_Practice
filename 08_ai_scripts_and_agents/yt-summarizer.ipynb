```python
import re
from fastmcp import FastMCP
from youtube_transcript_api import YouTubeTranscriptApi

# Initialize FastMCP
mcp = FastMCP("YouTube Summarizer")


def extract_video_id(url: str) -> str:
    """
    Extract the 11-character YouTube video ID from various URL formats.
    """
    pattern = r"(?:v=|\/shorts\/|\/embed\/|\/v\/|youtu\.be\/)([a-zA-Z0-9_-]{11})"
    match = re.search(pattern, url)
    return match.group(1) if match else None


@mcp.tool()
def get_youtube_transcript(video_url: str) -> str:
    """
    Fetch the transcript of a YouTube video.

    Args:
        video_url: Full YouTube video URL.

    Returns:
        Complete transcript text or an error message.
    """

    video_id = extract_video_id(video_url)

    if not video_id:
        return "Error: Invalid YouTube URL."

    try:
        # Create API instance (Required for youtube-transcript-api v1.x)
        api = YouTubeTranscriptApi()

        # Fetch transcript
        transcript = api.fetch(video_id)

        # Convert transcript snippets into one string
        full_text = " ".join(snippet.text for snippet in transcript)

        return full_text

    except Exception as e:
        return f"Error: Could not retrieve transcript.\nDetails: {str(e)}"


if __name__ == "__main__":
    mcp.run()
```
