```python
import chromadb

client = chromadb.PersistentClient(path="./My_chromaDB")
```

```python
from chromadb.utils import embedding_functions
```

```python
import chromadb.utils.embedding_functions as embedding_functions

# 2. Capitalize "MiniLM" exactly like this (notice the capital M's)
sentence_transformers_ef = embedding_functions.SentenceTransformerEmbeddingFunction(
    model_name="all-MiniLM-L6-v2"
)
```

```python
collection=client.create_collection(
    name="my_response4",
    embedding_function=sentence_transformers_ef,
    metadata={"description" : "my favourate recipes"},
)
```

```python

print(f"collection name: {collection.name}")
print(f"collection metadata: {collection.metadata}")
```

```python
recipes = [
    {
        "name": "Spaghetti Aglio Olio",
        "ingredients": [
            "200g spaghetti",
            "4 cloves garlic, thinly sliced",
            "1/2 tsp red pepper flakes",
            "4 tbsp olive oil",
            "Salt to taste"
        ],
        "instructions": [
            "Cook spaghetti in salted boiling water until al dente.",
            "Heat olive oil in a pan, add garlic and red pepper flakes, and cook until golden.",
            "Toss cooked spaghetti in the pan with a splash of pasta water, then serve."
        ]
    },
    {
        "name": "Classic Pancakes",
        "ingredients": [
            "1 cup all-purpose flour",
            "2 tbsp sugar",
            "1 tsp baking powder",
            "1 cup milk",
            "1 egg",
            "2 tbsp melted butter"
        ],
        "instructions": [
            "Whisk flour, sugar, and baking powder in a bowl.",
            "Mix in the milk, egg, and melted butter until smooth.",
            "Pour batter onto a hot greased griddle and cook until bubbles form, then flip until golden brown."
        ]
    }
]

# Example: Print the names of all recipes in the list
for recipe in recipes:
    print(f"Recipe: {recipe['name']}")
```

```python
recipes
```

```python
recipe_metadata = [
    {
        "name": "Spaghetti Aglio Olio",
        "prep_time_mins": 5,
        "cook_time_mins": 10,
        "servings": 2,
        "difficulty": "Easy",
        "calories": 450,
        "tags": ["pasta", "dinner", "quick", "vegetarian"]
    },
    {
        "name": "Classic Pancakes",
        "prep_time_mins": 10,
        "cook_time_mins": 15,
        "servings": 4,
        "difficulty": "Easy",
        "calories": 350,
        "tags": ["breakfast", "sweet", "kid-friendly"]
    }
]

# Example: Print metadata details for each recipe
for data in recipe_metadata:
    total_time = data["prep_time_mins"] + data["cook_time_mins"]
    print(f"{data['name']}: Takes {total_time} mins total. Level: {data['difficulty']}.")
```

```python
print(recipe_metadata)
```

```python
ids=["recipe_1","recipe_2"]
```

```python
# Convert recipe dictionaries to strings
recipes_as_strings = []
for recipe in recipes:
    recipe_text = f"Recipe: {recipe['name']}\n"
    recipe_text += f"Ingredients: {', '.join(recipe['ingredients'])}\n"
    recipe_text += f"Instructions: {' '.join(recipe['instructions'])}"
    recipes_as_strings.append(recipe_text)

# Now add to collection with string documents
collection.add(
    documents=recipes_as_strings,  # Use the converted strings
    metadatas=recipe_metadata,     # Keep metadata as dictionaries
    ids=ids
)
```

```python
all_data=collection.get()

print(all_data)
```
