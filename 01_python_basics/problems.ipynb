```python
# 1. Print Numbers
# Print numbers from 1 to 10 using a for loop.
for i in range(1, 11):
    print(i)
```

```python
# 2. Print Even Numbers
for i in range(1, 21):
    if i % 2 == 0:
        print(i)
```

```python
# 3. Print Odd Numbers
for i in range(1, 21):
    if i % 2 != 0:
        print(i)
```

```python
# 4. Sum of Numbers
# Find the sum of numbers from 1 to 100.

total_sum = 0
for i in range(1, 101):
    total_sum += i
print(total_sum)
```

```python
# 5. Multiplication Table
for i in range(1, 11):
    print(f"7 x {i} = {7 * i}")
```

```python
# 6. Count Characters
# Count the number of characters in a string without using len().
# Input: text = "Python"
text = "Python"
count = 0
for char in text:
    count += 1
print(count)
```

```python
# 7. Print Each Character
# Print each character of a string on a new line.
# Input: text = "Hello"
text = "Hello"
for char in text:
    print(char)
```

```python
# 8. Find Largest Number
# Find the largest number in a list using a loop.
# Input: nums = [10, 45, 23, 89, 12]
nums = [10, 45, 23, 89, 12]
largest = nums[0]
for num in nums:
    if num > largest:
        largest = num
print(largest)
```

```python
# 9. Count Vowels
# Count the number of vowels in a string.
# Input: text = "Artificial Intelligence"
text = "Artificial Intelligence"
vowels = "aeiouAEIOU"
vowel_count = 0
for char in text:
    if char in vowels:
        vowel_count += 1
print(vowel_count)
```

```python
# 10. Reverse a String
# Reverse a string using a loop.
# Input: text = "Python"
text = "Python"
reversed_text = ""
for char in text:
    reversed_text = char + reversed_text
print(reversed_text)
```
