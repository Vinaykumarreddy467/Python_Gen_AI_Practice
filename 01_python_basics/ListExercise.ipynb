```python
# ==============================================================================
# PART 1: BASIC LIST MANIPULATIONS (EXPRESSIONS & OUTPUTS)
# ==============================================================================

print("="*60 + "\nPART 1: BASIC LIST MANIPULATIONS\n" + "="*60)

# 1. Add 'orange' to the end of ['apple', 'banana', 'mango']
fruits = ['apple', 'banana', 'mango']
fruits.append("orange")
print(f"1. Append: {fruits}")

# 2. Insert 'kiwi' at index 1
fruits = ['apple', 'banana', 'mango']
fruits.insert(1, "kiwi")
print(f"2. Insert: {fruits}")

# 3. Remove 'banana' from the list
fruits = ['apple', 'banana', 'mango']
fruits.remove('banana')
print(f"3. Remove: {fruits}")

# 4. Remove the last item using pop()
fruits = ['apple', 'banana', 'mango']
popped_item = fruits.pop()
print(f"4. Pop last: {fruits} (Removed item: '{popped_item}')")

# 5. Find the total number of students
students_list = ['Rahul', 'Priya', 'Aman', 'Neha']
print(f"5. Total students: {len(students_list)}")

# 6. Count how many times 10 appears
numbers_list = [10, 20, 10, 30, 10, 40]
print(f"6. Occurrences of 10: {numbers_list.count(10)}")

# 7. Find the index of 'Bangalore'
city_list = ['Delhi', 'Mumbai', 'Bangalore', 'Chennai']
print(f"7. Index of 'Bangalore': {city_list.index('Bangalore')}")

# 8. Reverse the list [1, 2, 3, 4, 5]
numbers = [1, 2, 3, 4, 5]
numbers.reverse()  # or numbers[::-1]
print(f"8. Reversed list: {numbers}")

# 9. Sort the list [45, 10, 99, 23, 1]
unsorted_list = [45, 10, 99, 23, 1]
unsorted_list.sort()
print(f"9. Sorted list: {unsorted_list}")

# 10. Clear all items from a fruit list
clear_fruits = ['apple', 'banana', 'mango']
clear_fruits.clear()
print(f"10. Cleared list: {clear_fruits}")

# 11. Combine two lists using extend()
list1 = [1, 2, 3]
list2 = [4, 5, 6]
list1.extend(list2)
print(f"11. Extended list: {list1}")

# 12. Create a backup copy of a list using copy()
original = ['A', 'B', 'C']
backup = original.copy()
print(f"12. Backup copy: {backup}")

# 13. Add multiple fruits using extend()
more_fruits = ['apple', 'banana']
more_fruits.extend(["grapes", "kiwi", "papaya"])
print(f"13. Multi-extend: {more_fruits}")

# 14. Remove an item by index using pop(index)
pop_by_idx = ['zero', 'one', 'two', 'three']
removed_val = pop_by_idx.pop(2)
print(f"14. Pop index 2: {pop_by_idx} (Removed item: '{removed_val}')")

# 15. Find the largest number using a built-in function
large_nums = [4, 12, 99, 1, 35]
print(f"15. Largest number: {max(large_nums)}")

# 16. Predict the output of append() when adding a list inside another list
nested_target = [1, 2, 3]
nested_target.append([4, 5, 6])
print(f"16. Append nested list result: {nested_target}")

# 17. Predict the output of pop() and the modified list
pop_demo = [10, 20, 30]
returned_demo = pop_demo.pop()
print(f"17. Pop mutation: list is now {pop_demo}, returned element is {returned_demo}")

# 18. Find the second largest number in a list
dup_max_list = [10, 20, 40, 40, 30]
unique_nums = list(set(dup_max_list))
unique_nums.sort()
print(f"18. Second largest number: {unique_nums[-2]}")

# 19. Remove duplicate values from a list
duplicates_list = [1, 2, 2, 3, 4, 4, 5]
unique_list = list(set(duplicates_list))
print(f"19. Duplicates removed: {unique_list}")


# ==============================================================================
# PART 2: CHAINED OPERATIONS CHALLENGE
# ==============================================================================

print("\n" + "="*60 + "\nPART 2: CHAINED OPERATIONS CHALLENGE\n" + "="*60)

challenge_nums = [10, 20, 30, 40, 50]
challenge_nums.append(60)
challenge_nums.insert(1, 15)
challenge_nums.remove(30)
challenge_nums.reverse()
challenge_nums.sort()
idx_40 = challenge_nums.index(40)
count_20 = challenge_nums.count(20)

print(f"Final mutated list: {challenge_nums}")
print(f"Index of 40: {idx_40}")
print(f"Count of 20: {count_20}")


# ==============================================================================
# PART 3: STUDENT MANAGEMENT SYSTEM
# ==============================================================================

print("\n" + "="*60 + "\nPART 3: STUDENT MANAGEMENT SYSTEM\n" + "="*60)

students = []

def display_menu():
    print("\n" + "="*40)
    print("  STUDENT MANAGEMENT SYSTEM  ")
    print("="*40)
    print("1. Add New Student")
    print("2. View All Students")
    print("3. Search Student by ID")
    print("4. Exit")
    print("="*40)

def add_student():
    student_id = input("Enter student ID: ")
    name = input("Enter student name: ")
    dept = input("Enter student department: ")
    contact = input("Enter student contact: ")
    students.append([student_id, name, dept, contact])
    print(f"Student '{name}' added successfully!")

def view_students():
    if not students:
        print("No student records found.")
        return
    print("\nID\t| Name\t\t| Department\t| Contact")
    print("-" * 50)
    for s in students:
        print(f"{s[0]}\t| {s[1]}\t\t| {s[2]}\t\t| {s[3]}")

def search_student():
    search_id = input("Enter Student ID to search: ")
    for s in students:
        if s[0] == search_id:
            print("\nStudent Found!")
            print(f"ID: {s[0]}\nName: {s[1]}\nDept: {s[2]}\nContact: {s[3]}")
            return
    print("Student ID not found.")

# Un-comment the block below to run the interactive system if executing locally
"""
while True:
    display_menu()
    choice = input("Enter your Choice: ")
    match choice:
        case "1": add_student()
        case "2": view_students()
        case "3": search_student()
        case "4": 
            print("Exiting system. Goodbye!")
            break
        case _: print("Invalid choice. Select 1-4.")
"""
print("Student Management System code loaded. (Main loop commented out to prevent execution hang)")
```
